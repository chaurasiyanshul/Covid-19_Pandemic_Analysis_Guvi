# Covid-19 Pandemic Total cases and Death analysis

```python
import pandas as pd

df = pd.read_csv("day_wise.csv")
df.head()
```

```python
df.info()
```

#Line plot between date and daily cases

```python
import plotly.express as px

fig = px.line(df, x="Date", y="Confirmed", title="Daily Cases Over Time")
fig.add_scatter(x=df['Date'], y=df['Active'], mode='lines', name='Active Cases')
fig.add_scatter(x=df['Date'], y=df['Deaths'], mode='lines', name='Deaths')
fig.add_scatter(x=df['Date'], y=df['Recovered'], mode='lines', name='Recovered')
fig.show()
```

```python
df1 = pd.read_csv("country_wise_latest.csv")
df1.head()
```

#Total Deaths

```python
total_deaths = df['Deaths'].sum()
print("Total Deaths:", total_deaths)
```

# Total Recovered

```python
total_recovered = df['Recovered'].sum()
print("Total Recovered:", total_recovered)
```

# Total cofirmed Cases

```python
total_cases = df['Confirmed'].sum()
print("Total Cases:", total_cases)
```

# Bar chart: Total cases by country

```python
import plotly.express as px

fig = px.bar(df1, y="Confirmed", x="Country/Region", title="Total Cases by Country")
fig.show()
```

# Pie chart: Recovery vs Death rate

```python
import plotly.express as px

# Calculate total recovered and total deaths from df1
total_recovered = df1['Recovered'].sum()
total_deaths = df1['Deaths'].sum()

# Create a dictionary for the pie chart data
data = {'Category': ['Recovered', 'Deaths'],
        'Count': [total_recovered, total_deaths]}
pie_df = pd.DataFrame(data)

# Create the interactive pie chart
fig = px.pie(pie_df, values='Count', names='Category', title='Overall Recovery vs Death Rate')
fig.show()
```
